```python
import numpy as np
import random
import matplotlib.pyplot as plt
from pymoo.core.problem import ElementwiseProblem
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from pymoo.termination import get_termination

# --- Problem Parameters ---
N = 10  # Number of PEVGs (Electric Vehicles)
C = 99  # Grid capacity (MWh)

# Battery capacities (MWh) and Satisfaction coefficients
b_n = np.array([random.randint(35, 65) for _ in range(N)])
s_n = np.array([random.uniform(1, 2) for _ in range(N)])

# --- Define the Multi-Objective EV Charging Problem ---
class EVChargingProblem(ElementwiseProblem):
    def __init__(self):
        super().__init__(n_var=N + 1,  # p (price) + x_n (demand for each EV)
                         n_obj=2,      # 2 Objectives: Revenue, Utility
                         n_constr=1,   # 1 Constraint: Total demand <= C
                         xl=np.array([20] + [0] * N),  # Lower bounds (Price ≥ 20, Demand ≥ 0)
                         xu=np.array([50] + [min(b_n[i], C) for i in range(N)]))  # Upper bounds (Price ≤ 50, Demand ≤ b_n)

    def _evaluate(self, X, out, *args, **kwargs):
        p = X[0]   # Charging price
        x_n = X[1:]  # Charging demand for each user

        # --- Objective 1: Maximize Revenue ---
        revenue = p * np.sum(x_n)  

        # --- Objective 2: Maximize User Utility ---
        utility = np.sum(b_n * x_n - 0.5 * s_n * x_n**2 - p * x_n)

        # --- Constraint: Total Charging Demand <= Grid Capacity ---
        wastage = C - np.sum(x_n)  

        # Convert maximization to minimization for pymoo
        out["F"] = [-revenue, -utility]  
        out["G"] = [-wastage]  # Constraint as g(x) ≤ 0

# --- Initialize NSGA-II Algorithm ---
algorithm = NSGA2(
    pop_size=100,
    n_offsprings=50
)

# --- Termination Criteria ---
termination = get_termination("n_gen", 150)
```

```python
# --- Solve the Problem ---
res = minimize(EVChargingProblem(), algorithm, termination, seed=42, verbose=False , save_history=True)

# --- Extract Results ---
X = res.X  # Decision variables (Price & Demand per user)
F = res.F  # Objective values (negated Revenue & Utility)

# Convert back to maximization
revenue = -F[:, 0]  
utility = -F[:, 1]  

# --- PARETO FRONT PLOT ---
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(revenue, utility, c='blue', label='Pareto Front')
ax.set_xlabel('Revenue (USD)')
ax.set_ylabel('Utility')
ax.set_title('Pareto Front: Revenue vs. Utility')
plt.legend()
plt.savefig('pareto_front_ev_charging.jpg')
plt.show()
```

```python
len(res.history[0].pop.get("F"))
```

```python
pareto_front = np.column_stack((revenue, utility))
```

```python
best_distances = []

for gen in res.history:
    gen_F = -gen.pop.get("F")  # Get objective values in this generation

    # Compute minimum distance from each solution in the current generation to the Pareto front
    distances = [np.min(np.linalg.norm(pf - gen_F, axis=1)) for pf in pareto_front]
    
    # Track the overall minimum distance at this generation
    best_distances.append(np.min(distances))

# --- Plot Convergence ---
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(best_distances, label="Convergence (Distance to Pareto Front)", color='red')
ax.set_xlabel('Generations')
ax.set_ylabel('Distance to Pareto Front')
ax.set_title('Convergence of NSGA-II')
plt.legend()
plt.savefig('convergence_ev_charging.jpg')
plt.show()
```
