```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Problem parameters
N = 5  # Number of PEVGs
C = 99  # Grid capacity (MWh)
b_n = np.random.uniform(35, 65, N)  # Battery capacities
s_n = np.random.uniform(1, 2, N)    # Satisfaction parameters
```

```python
# PSO parameters
n_particles = 30
n_iterations = 100
w = 0.7    # Inertia weight
c1 = 1.5   # Cognitive coefficient
c2 = 1.5   # Social coefficient
p_bounds = (0, 50)  # Price bounds
x_bounds = (0, C)   # Demand bounds
```

```python
# Fitness function: Optimize PEVG utility, compute SG revenue
def fitness(solution):
    p = solution[0]
    x = solution[1:]
    
    # PEVG total utility (primary objective)
    total_demand = np.sum(x)
    pevg_utility = np.sum(b_n * x - 0.5 * s_n * x**2 - p * x)
    
    # SG revenue (secondary computation)
    sg_revenue = p * total_demand
    
    # Penalty for constraint violation
    penalty = 0
    if total_demand > C:
        penalty = 1e6 * (total_demand - C)
    
    return pevg_utility - penalty, sg_revenue
```

```python
# PSO algorithm
def pso_optimize():
    # Initialize particles: [p, x_1, x_2, ..., x_N]
    particles = np.zeros((n_particles, N + 1))
    particles[:, 0] = np.random.uniform(p_bounds[0], p_bounds[1], n_particles)  # p
    particles[:, 1:] = np.random.uniform(x_bounds[0], x_bounds[1], (n_particles, N))  # x_n
    
    velocities = np.zeros((n_particles, N + 1))
    pbest = particles.copy()
    pbest_fitness = np.array([fitness(particle)[0] for particle in particles])  # Only PEVG utility for optimization
    pbest_sg_revenue = np.array([fitness(particle)[1] for particle in particles])  # Track SG revenue
    gbest_idx = np.argmax(pbest_fitness)
    gbest = pbest[gbest_idx]
    gbest_fitness = pbest_fitness[gbest_idx]
    gbest_sg_revenue = pbest_sg_revenue[gbest_idx]

    # History for convergence
    pevg_history = []
    sg_history = []

    for _ in range(n_iterations):
        r1, r2 = np.random.rand(n_particles, N + 1), np.random.rand(n_particles, N + 1)
        velocities = (w * velocities + 
                      c1 * r1 * (pbest - particles) + 
                      c2 * r2 * (gbest - particles))
        particles = particles + velocities
        
        # Enforce bounds
        particles[:, 0] = np.clip(particles[:, 0], p_bounds[0], p_bounds[1])
        particles[:, 1:] = np.clip(particles[:, 1:], x_bounds[0], x_bounds[1])
        
        # Evaluate fitness
        current_fitness = np.array([fitness(particle) for particle in particles])  # Returns (pevg_utility, sg_revenue)
        current_pevg_fitness = current_fitness[:, 0]
        current_sg_revenue = current_fitness[:, 1]
        
        # Update personal best (based on PEVG utility)
        improved = current_pevg_fitness > pbest_fitness
        pbest[improved] = particles[improved]
        pbest_fitness[improved] = current_pevg_fitness[improved]
        pbest_sg_revenue[improved] = current_sg_revenue[improved]
        
        # Update global best
        if np.max(current_pevg_fitness) > gbest_fitness:
            gbest_idx = np.argmax(current_pevg_fitness)
            gbest = particles[gbest_idx]
            gbest_fitness = current_pevg_fitness[gbest_idx]
            gbest_sg_revenue = current_sg_revenue[gbest_idx]
        
        pevg_history.append(gbest_fitness)
        sg_history.append(gbest_sg_revenue)
    
    return gbest, gbest_fitness, gbest_sg_revenue, pevg_history, sg_history
```

```python
# Run PSO
solution_opt, pevg_utility_opt, sg_revenue_opt, pevg_history, sg_history = pso_optimize()
p_opt = solution_opt[0]
x_opt = solution_opt[1:]
total_demand = np.sum(x_opt)
```

```python
# Post-process: Adjust p to maximize SG revenue (Stackelberg equilibrium)
p_star = np.min(b_n - s_n * x_opt)  # p* = min(b_n - s_n x_n^*) from paper
sg_revenue_star = p_star * total_demand
pevg_utility_star = np.sum(b_n * x_opt - 0.5 * s_n * x_opt**2 - p_star * x_opt)
```

```python
# Output results
print("PSO Results (Optimizing PEVG Utility):")
print(f"Optimal Price (p): {p_opt:.2f} USD/MWh")
print(f"Optimal Demands (x_n): {x_opt}")
print(f"Total Demand: {total_demand:.2f} MWh (Constraint: <= {C})")
print(f"PEVG Total Utility: {pevg_utility_opt:.2f}")
print(f"SG Revenue: {sg_revenue_opt:.2f}")
print()
```

```python

print("Post-Processed Stackelberg Equilibrium:")
print(f"Adjusted Price (p*): {p_star:.2f} USD/MWh")
print(f"Fixed Demands (x_n): {x_opt}")
print(f"SG Revenue: {sg_revenue_star:.2f}")
print(f"PEVG Total Utility (with p*): {pevg_utility_star:.2f}")
```

```python
# Plot convergence
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(pevg_history, label="PEVG Utility")
plt.xlabel("Iteration")
plt.ylabel("PEVG Total Utility")
plt.title("PEVG Utility Convergence")
plt.legend()
plt.grid()
```

```python
plt.subplot(1, 2, 2)
plt.plot(sg_history, label="SG Revenue")
plt.xlabel("Iteration")
plt.ylabel("SG Revenue")
plt.title("SG Revenue Convergence")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
